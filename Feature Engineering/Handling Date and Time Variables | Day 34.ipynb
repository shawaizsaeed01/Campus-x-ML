# Working with Dates

```python
# converting to datetime to datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1. Extract year

```python
# extracting year from a date
date['date_year'] = date['date'].dt.year
date.sample(5)
```

### 2. Extract Month

```python
# extracting month from a date
date['date_month'] = date['date'].dt.month
date.sample(5)
```

```python
# extracting month_name from a date
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

### Extract Days

```python
# extracting days from a days
date['date_day'] = date['date'].dt.day
date.sample(5)
```

```python
# extracting day of week
date['date_day_of_week'] = date['date'].dt.dayofweek
date.sample(5)
```

```python
# extracting day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# extracting is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract week of the year

```python
import warnings
warnings.filterwarnings("ignore")

date['date_week'] = date['date'].dt.week

date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```
